```python
%matplotlib inline
from getdist import MCSamples, plots
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('mf_hmf_cov_clean.txt', usecols=(0, 1))
samples = MCSamples(samples=data, names=['omega_m', 'sigma8_input'], labels=[r'\Omega_m', r'\sigma_8'])
g = plots.get_subplot_plotter()
g.triangle_plot(samples, filled=True)
g.plot_2d(samples, 'omega_m', 'sigma8_input', filled=True)

# Optional: plot parameter traces manually
fig, axes = plt.subplots(2, 1, figsize=(8, 4), sharex=True)
axes[0].plot(data[:, 0])
axes[0].set_ylabel(r'$\Omega_m$')
axes[1].plot(data[:, 1])
axes[1].set_ylabel(r'$\sigma_8$')
axes[1].set_xlabel('Sample')
plt.tight_layout()
plt.show()
```
